```python
import requests as rq
import bs4
from bs4 import BeautifulSoup as bs
import re
import pandas as pd
```

```python
def golearn(i): # i = number of pages
  Matn = []
  Name = []
  Tag = []
  Like = []
  for j in range(1,i+1):
    url = 'https://www.goodreads.com/quotes/tag/science?page=' + str(j)
    site = rq.get(url)
    soup = bs(site.text , 'html.parser')
    box = soup.find_all('div', attrs = {'class':'quote mediumText '})
    matn = []
    tag = []
    like = []
    for i in range(0,len(box)):
      q = box[i].find('div', attrs = {'class':'quoteText'}).text.replace('\n','').strip()
      tagg = box[i].find('div', attrs = {'class':'greyText smallText left'}).text.replace('\n','').strip()
      likee = box[i].find('div', attrs = {'class':'right'}).text.replace('\n','').strip()
      matn.append(q)
      tag.append(tagg)
      like.append(likee)
    matnn = []
    name = []
    for i in range(0,len(matn)):
      w = matn[i] + ','
      w1 = re.findall('“(.+?)\”',w)
      w2 = re.findall('”(.*?)\,',w)
      w3 = re.findall('\w.*\w',w2[0])
      matnn.append(w1[0])
      name.append(w3[0].strip())
    e = []
    for i in range(0,len(tag)):
      t = tag[i].split(':')
      t = t[1].split(',')
      for i in range(0,len(t)):
        t[i] = t[i].strip()
      e.append(t)
    likee = []
    for i in range(0,len(like)):
      s = like[i].split(' ')
      likee.append(s[0])
    Matn = Matn + matnn
    Name = Name + name
    Tag = Tag + e
    Like = Like + likee
  return Matn , Name, Tag, Like
```

```python
a,b,c,d = golearn(50)
```

```python
C = []

for i in range(0,len(c)):
  r = c[i]
  s = ''
  for i in range (0,len(r)):
    if i == len(r) - 1:
      s += r[i] 
    else:
      s += r[i] + '|'
  C.append(s)
```

```python
rows = []
for i in range(0,len(a)):
  A = [a[i],b[i],C[i],d[i]]
  rows.append(A)
```

```python
import csv 
    

fields = ['quote', 'name','tag' ,'likes']

rows = rows 

filename = "phase1_.csv"

with open(filename, 'w') as csvfile: 
    # creating a csv writer object 
    csvwriter = csv.writer(csvfile) 
        
    # writing the fields 
    csvwriter.writerow(fields) 
        
    # writing the data rows 
    csvwriter.writerows(rows)
```
