```python
import pandas as pd
import nltk
import numpy as np
```

```python
nltk.download('all')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import TweetTokenizer as TT
```

```python
data = pd.read_csv('/content/drive/MyDrive/data/goalearn/phase1_.csv')
```

```python
data
```

```python
x = data['quote']
```

```python
# creating the whole text
T = x[0]
for i in range(1,len(x)):
  T += x[i]
T
```

```python
T[63:66]
```

```python
import re
# finding words with ' (like I'm, you'r, ...)
T2 = re.findall('\w+\'\w',T) # notice that after ' comes just one letter

T2
```

```python
# finding numbers
T3 = re.findall('\d+',T)

T3
```

```python
# finding words
T4 = re.findall('\w{2,}',T)
```

```python
T4
```

```python
tk = TT()
token = tk.tokenize(T)
```

```python
token[100:200]
```

```python
# deleting words with '
en = T2

text1 = []
for word in token:
  if word.lower() not in en:
    text1.append(word)
```

```python
# deleting numbers
en = T3

text2 = []
for word in text1:
  if word.lower() not in en:
    text2.append(word)
```

```python
# gathering (just) words
en = T4

text3 = []
for word in text2:
  if word.lower() in en:
    text3.append(word)
```

```python
text3
```

```python
# deleting non-important words
en = stopwords.words('english')

text4 = []
for word in text3:
  if word.lower() not in en:
    text4.append(word)
```

```python
text4
```

```python
# making a complete text (notice that text4 is tokenized)
R = text4[0]
for i in range(1,len(text3)):
  R += ' ' + text3[i]
```

```python
R
```

now, R is the cleaned and text4 is the tokenized cleaned data!

```python
from wordcloud import WordCloud
# Generate a word cloud image
wordcloud = WordCloud(collocations = False,stopwords=en).generate(R)

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.imshow(wordcloud)
plt.axis("off")
```

```python
name = []
count = []
for i in range(0,len(text4)):
  word = text4[i]
  if word not in name:
    name.append(word)
    count.append(1)
  else:
    arg = name.index(word)
    count[arg] += 1
```

```python
count1 = np.array(count)
name1 = np.array(name)
name2 = []
for i in range(0,len(count1)):
  argm = np.argmax(count1)
  nam = name1[argm]
  name2.append(nam)
  count1 = np.delete(count1, argm)
  name1 = np.delete(name1, argm)
```

```python
count = np.array(count)
```

```python
d = {'word': name2, 'count': -np.sort(-count)}
result = pd.DataFrame(data=d)
result
```
